```python
# Import the dependencies.
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Create a set of random latitude and longitude combinations.
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)
lat_lngs
```

```python
# Create a practice set of random latitude and longitude combinations into a lists and pack them into the zipped tuple
x = [25.12903645, 25.92017388, 26.62509167, -59.98969384, 37.30571269]
y = [-67.59741259, 11.09532135, 74.84233102, -76.89176677, -61.13376282]
coordinates = zip(x, y)
```

```python
# Use the tuple() function to display the latitude and longitude combinations.
for coordinate in coordinates:
    print(coordinate[0], coordinate[1])
```

```python
# Add the latitudes and longitudes to a list.
## You can only unzip a zipped tuple once before it is removed from the computer's memory. Make sure you unzip the 
## latitudes and longitudes into the coordinates list before moving on.
coordinates = list(lat_lngs)
```

```python
# Use the citipy module to determine city based on latitude and longitude.
from citipy import citipy
```

```python
# Use the tuple() function to display the latitude and longitude combinations.
# Iterate through the coordinates' zipped tuple.
# Use citipy.nearest_city() and inside the parentheses of nearest_city(), add the latitude and longitude in this format: coordinate[0], coordinate[1].
# To print the city name, chain the city_name to the nearest_city() function.
# To print the country name, chain the country_code to the nearest_city() function.

for coordinate in coordinates:
    print(citipy.nearest_city(coordinate[0], coordinate[1]).city_name,
          citipy.nearest_city(coordinate[0], coordinate[1]).country_code)
```

```python
# Create an endpoint URL for a city.
city_url = url + "&q=" + "Boston"
city_weather = requests.get(city_url)
city_weather.json()
```

```python
# Get the JSON data.
boston_data = city_weather.json()
```

```python
boston_data["sys"]["country"]
```

```python
boston_data["main"]
```

```python
lat = boston_data["coord"]["lat"]
lng = boston_data["coord"]["lon"]
max_temp = boston_data["main"]["temp_max"]
humidity = boston_data["main"]["humidity"]
clouds = boston_data["clouds"]["all"]
wind = boston_data["wind"]["speed"]
print(lat, lng, max_temp, humidity, clouds, wind)
```

```python
# Import the datetime module from the datetime library.
from datetime import datetime
# Get the date from the JSON file.
date = boston_data["dt"]
# Convert the UTC date to a date format (ISO) with year, month, day, hours, minutes, and seconds.
datetime.utcfromtimestamp(date)
```

```python
#readable date format
datetime.utcfromtimestamp(date).strftime('%Y-%m-%d %H:%M:%S')
```
